 Project Title: Exploratory Data Analysis (EDA) on Retail Sales Data (Dataset 1)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

 # 1. Data Loading and Cleaning: Load the retail sales dataset.
df1= pd.read_csv("retail_sales_dataset.csv")
df1

df1.isnull().sum()

df1['Date'] = pd.to_datetime(df1['Date'])
df1['Month'] = df1['Date'].dt.to_period('M')
df1

# 2. Descriptive Statistics: Calculate basic statistics (mean, median, mode, standard deviation)

df1.describe()
print("Mean of Total Amount:", df1['Total Amount'].mean())
print("Median of Total Amount:", df1['Total Amount'].median())
print("Standard Deviation of Total Amount:", df1['Total Amount'].std())
print("Most Common total Amount:", df1['Total Amount'].mode()[0])
print("Most Common Quantity:", df1['Quantity'].mode()[0])

# 3. Time Series Analysis: Analyze sales trends over time using time series techniques.
monthly_sales = df1.groupby('Month')['Total Amount'].sum()
monthly_sales.plot(kind='line', marker='o', figsize=(10,5), title='Monthly Sales Trend')
plt.xlabel('Months')
plt.ylabel('Total Sales Amount')
plt.grid()
plt.show()

# 4. Customer and Product Analysis: Analyze customer demographics and purchasing behavior.

# Product Analysis - Purchasing Behavior
products = df1.groupby('Product Category')['Total Amount'].sum()
products.plot(kind='bar',color='skyblue',title='Product Categories by Sales')
plt.xlabel("Product Category")
plt.ylabel("Total Sales (sum of Total Amount)")
plt.show()

# Customer Analysis - customer demographics 
Genders = df1.groupby("Gender")['Total Amount'].sum()
Genders.plot(kind='pie',autopct='%1.1f%%', startangle=90, title='Sales by Gender (Customer Demographics)')
plt.ylabel('')
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df1['Age'], bins=10, kde=True, color='Purple')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Customers')
plt.show()

# 5. Visualization: Present insights through bar charts, line plots, and heatmaps.

plt.figure(figsize=(9,6))
sns.heatmap(df1.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```
